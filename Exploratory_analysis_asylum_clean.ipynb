```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
asylum_data = pd.read_csv('./asylum_clean.csv')
```

```python
summary = asylum_data.describe()
```

```python
value_counts = {}
for attr in asylum_data:
    value_counts[attr] =  asylum_data[attr].nunique()

value_counts = {'Unique_vals':value_counts}
value_counts = pd.DataFrame.from_dict(value_counts).T

frames = [summary, value_counts]

final = pd.concat(frames)
final
```

```python
granted = asylum_data.loc[asylum_data.grantraw==1, :]
denied = asylum_data.loc[asylum_data.grantraw==0, :]
```

```python
%matplotlib inline

def make_hist(feature):
    try:
        plt.figure()
        n, bins, patches = plt.hist(granted[feature].dropna().values, color ='b', alpha=0.5, normed=True, label='Granted')
        plt.hist(denied[feature].dropna().values, bins, color ='y', alpha=0.5, normed=True, label='Denied')
        plt.xlabel(str(feature))
        plt.ylabel('Frequency (normed)')
        plt.legend()
    except:
        return feature
```

```python
bad_features = []
for feature in asylum_data:
    bad_features.append(make_hist(feature))
```
